```python
def split_addrs(s): 
    #split an address list into list of tuples of (name,address) 
    if not(s): return [] 
    outQ = True 
    cut = -1 
    res = [] 
    for i in range(len(s)): 
        if s[i]=='"': outQ = not(outQ) 
        if outQ and s[i]==',': 
            res.append(email.utils.parseaddr(s[cut+1:i])) 
            cut=i 
    res.append(email.utils.parseaddr(s[cut+1:i+1])) 
    return res 
```
